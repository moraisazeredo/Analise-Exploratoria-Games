```python
# Bibliotecas para tratar dados
import pandas as pd
import numpy as np

# Bibliotecas para EDA
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

from plotly.subplots import make_subplots

import warnings

warnings.filterwarnings("ignore")

plt.style.use("ggplot")

pd.set_option('display.max_rows', 15)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('data/vgsales.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna()
```

Para começar, vamos fazer uma análise univariada nas nossas variáveis:

```python
fig = px.histogram(df, x='Genre')

fig.update_xaxes(categoryorder='total descending')

fig.update_layout(title='Distribuição de gêneros dos jogos',
                 xaxis_title='Gênero',
                 yaxis_title='Contagem')
fig.show()
```

```python
fig = px.histogram(df, x='Platform')

fig.update_xaxes(categoryorder='total descending')

fig.update_layout(title='Distribuição de Plataformas',
                 xaxis_title='Platform',
                 yaxis_title='Contagem')

fig.show()
```

Qual ano teve o maior número de lançamentos?

```python
plt.figure(figsize=(15, 10))
sns.countplot(x="Year", data=df, order = df.groupby(by=['Year'])['Name'].count().sort_values(ascending=False).index)
plt.xticks(rotation=90)

plt.show()
```

Qual ano teve o maior número de vendas globais?

```python
data_sales = df.groupby('Year')['Global_Sales'].sum().reset_index()

plt.figure(figsize=(10,5))
sns.barplot(x='Year',
           y='Global_Sales',
           data=data_sales)

plt.xticks(rotation=90)
plt.xlabel('Ano')
plt.ylabel('Vendas Globais')

plt.show()
```

Agora que nós conhecemos melhor sobre os nossos dados vamos fazer análises mais profundas com mais variáveis. Vamos olhar a relação entre os gêneros e plataformas e seus preços globais:

```python
genre_sales = df.groupby('Genre')['Global_Sales'].sum()
genre_sales = genre_sales.reset_index()
genre_sales = genre_sales.sort_values('Global_Sales', ascending=False)

platform_sales = df.groupby('Platform')['Global_Sales'].sum()
platform_sales = platform_sales.reset_index()
platform_sales = platform_sales.sort_values('Global_Sales', ascending=False)

fig = make_subplots(rows=1, cols=2)

fig.add_trace(
    go.Bar(x=genre_sales['Genre'],
            y=genre_sales['Global_Sales'],
            name='Vendas Globais por Gênero'),
            row=1, col=1)

fig.add_trace(
    go.Bar(x=platform_sales['Platform'],
        y=platform_sales['Global_Sales'],
        name='Vendas Globais por Plataforma'),
        row=1, col=2)

fig.update_layout(title='Vendas Globais por Gênero e Plataforma')
fig.show()
```

Repare que o gênero Shooter era o 5º na contagem de jogos lançados, mas seu faturamento global fica dentro do top 3.

Como esperado, o PS2 foi a plataforma que mais vendeu jogos. Isso pode ter ocorrido pois o PS2 foi uma plataforma muito famosa no seu lançamento, devido ao alto avanço gráfico e de jogabilidade em comparação com o as plataformas anteriores.

Repare que o gênero mais vendido no mundo foi ação, porém, será que o jogo mais vendido no mundo é necessariamente um jogo de ação? Vamos analisar qual foi o jogo mais vendido no mundo:

```python
top_game_sale = df.head(20)
top_game_sale = top_game_sale.sort_values(by=['Global_Sales'], ascending=False)
```

```python
name = top_game_sale['Name']
year = top_game_sale['Year']
y = np.arange(0, 20)
```

```python
plt.figure(figsize=(30, 18))
g = sns.barplot(x='Name', y='Global_Sales', data=top_game_sale)
index = 0

for value in top_game_sale['Global_Sales']:
    g.text(index, value - 18, name[index], color='#000', size=14, rotation= 90, ha="center")
    index += 1

plt.xticks(y, top_game_sale['Year'], fontsize=14, rotation=90)
plt.xlabel('Ano de Lançamento')
plt.ylabel('Vendas Globais')

plt.show()
```

E se olharmos para as vendas por região, quais são os gêneros mais vendidos?

```python
comp_map = df.groupby(by=['Genre']).sum()
comp_table = comp_map.reset_index()
comp_table = pd.melt(comp_table, id_vars=['Genre'], value_vars=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], 
                    var_name='Sale_Area', value_name='Sale_Price')
comp_table
```

```python
# Crie o gráfico de barras agrupado com o Plotly
fig = px.bar(comp_table, x='Genre', y='Sale_Price', color='Sale_Area',
             title='Vendas regionais por gênero',
             barmode='group')  # Defina o modo de agrupamento

# Personalize o layout (opcional)
fig.update_layout(
    xaxis_title='Gênero',
    yaxis_title='Vendas Globais',
    legend_title='Área de Vendas'
)

# Exiba o gráfico
fig.show()
```

Repare que os Estados Unidos são os maiores consumidores de todas as categorias de games. 

Vamos olhar quais são as empresas que mais publicam games:

```python
# Agrupe os dados e conte o número de jogos por editora
publish_count = df.groupby('Publisher')['Name'].count().sort_values(ascending=False)[:10]

# Crie o gráfico de barras com o Plotly
fig = px.bar(x=publish_count.index, 
             y=publish_count.values,
             title='Quantidade de Jogos por Editora',
             text=publish_count.values)

# Personalize o layout (opcional)
fig.update_layout(
    xaxis_tickangle=-45,  # Rotacione os rótulos do eixo x
    xaxis_title='Editora',
    yaxis_title='Quantidade de Jogos'
)

# Exiba o gráfico
fig.show()
```

Interessante... A Nintendo não está nem nos top 5 empresas que mais publicam games, mas os top 4 games mais vendidos no mundo são da Nintendo.


Embora a Nintendo tenha os jogos mais vendidos, isso não significa necessariamente que ela seja a editora que mais publica jogos.
A Nintendo pode focar em lançar títulos de alta qualidade e com grande apelo ao público, enquanto outras editoras podem optar por lançar uma variedade maior de jogos, incluindo títulos menores e experimentais.<br>

```python
vendas = df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False)[:10].reset_index()

fig = px.bar(vendas,
                x='Publisher',
                y='Global_Sales',
                title='Relação entre quantidade de jogos e vendas globais',
                color='Global_Sales')

fig.update_layout(
    xaxis_title='Quantidade de Jogos',
    yaxis_title='Vendas Globais',
    legend_title='Editora'
)

fig.show()
```

Como esperado, a Nintendo foi a empresa que mais faturou, pois seus jogos estão nos top mais vendidos no mundo. Isso significa que a Nintendo foca na qualidade do jogo, buscando aquilo que o usuário quer e não na quantidade de jogos.
